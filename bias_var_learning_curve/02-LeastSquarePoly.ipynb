```python
import numpy as np
import BiasVarv2 as bv
import matplotlib.pyplot as plt
import random
import seaborn as sns
import sys
sys.path.append("..")
from LeastSquare import LeastSquare as LS
from LinearRegression2 import LinearRegression2 as LR


from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

sns.set_style("whitegrid")
sns.set_palette(sns.color_palette("Set2"))
sns.set_context('notebook', font_scale=1.25, rc={"lines.linewidth": 3})
```

```python
f = lambda x: x**2 + 0.25*x + 0.5
fx = 'x**2 + 0.25*x + 0.5'
```

```python
n_res = 1000
x = np.linspace(-3, 3, n_res)
y = f(x)
```

```python
num_sample = 10
x_sample = np.random.choice(x, num_sample)
y_sample = f(x_sample) + np.random.randn(num_sample)
```

```python
plt.plot(x, y)
plt.scatter(x_sample, y_sample, color='blue')
```

```python
def create_degree_x(raw_x, degree):
    new_x = []
    for w in raw_x:
        xx = []
        for i in range(degree+1):
            xx.append(w**i)
        new_x.append(xx)
    return np.array(new_x)
```

```python
for i in range(2, 8):
    x_sample_poly = create_degree_x(x_sample, degree=i)
    ls = LS(x_sample_poly, y_sample)
    model_w = ls.parameters
    x_poly = create_degree_x(x, degree=i)
    predicted_y = x_poly.dot(model_w.T)
    
    fig = plt.figure()
    sns.lineplot(x, y, alpha=0.5, label=fx)
    sns.scatterplot(x_sample, y_sample, color='blue')
    sns.lineplot(x, predicted_y, label=f'degree: {i}', alpha=0.3)
    plt.axis([-3.5, 3.5, -2, 15])
```
