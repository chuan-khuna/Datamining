```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

sns.set_style("whitegrid")
colors = sns.color_palette("muted")
sns.set_palette(colors)
sns.set_context("paper", font_scale=1.25, rc={"lines.linewidth": 1.5})
```

```python
df = pd.read_csv('Seed_Data.csv')
df.head()
```

```python
standardize = lambda x: (x-x.mean()) / x.std()

X = df.iloc[:, 0:-1]
X = standardize(X)
Y = df['target']
```

```python
sns.pairplot(X)
```

```python
row_colors = Y.map(dict(zip(Y.unique(), colors)))
methods = ["single", "complete", "average", "weighted", "centroid", "median", "ward"]
# method 0 to 6
method = 2
clustermap = sns.clustermap(X, row_colors=row_colors, method=methods[method], cmap="Spectral")
plt.suptitle(f"Seed Dataset, clustering method: {methods[method]}", fontsize=16)
```
