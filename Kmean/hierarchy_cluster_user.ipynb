```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

sns.set_style("whitegrid")
colors = sns.color_palette("muted")
sns.set_palette(colors)
sns.set_context("paper", font_scale=1.25, rc={"lines.linewidth": 1.5})
```

```python
df = pd.read_csv('user_knowledge.csv')
df.head()
```

```python
X = df.iloc[:, 0:-1]
Y = df['UNS']
```

```python
fig = plt.figure(figsize=(8, 5), dpi=150)
sns.pairplot(df, hue='UNS')
# plt.savefig('uns pairplot.jpg', dpi=200)
```

```python
row_colors = Y.map(dict(zip(Y.unique(), colors)))
methods = ["single", "complete", "average", "weighted", "centroid", "median", "ward"]
# method 0 to 6
method = 2
clustermap = sns.clustermap(X[['PEG']], row_colors=row_colors, method=methods[method], cmap="mako", col_cluster=False)
plt.suptitle(f"User Knowledge, clustering method: {methods[method]}", fontsize=16)
# clustermap.savefig("usrknowledge.jpg", dpi=300)
```
